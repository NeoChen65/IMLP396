## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
model = LinearRegression()
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, y_train_pred, color='red')
plt.legend()
plt.show()
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
x_test_value = np.array([2.44897959]).reshape(-1, 1)
y_test_pred_value = model.predict(x_test_value)
print(f"X = {x_test_value[0][0]:.8f}, 预测值 = {y_test_pred_value[0]:.8f}")
```

```python
from sklearn.metrics import mean_squared_error
x_test_reshaped = x_test.reshape(-1, 1)
y_test_pred = model.predict(x_test_reshaped)
mse = mean_squared_error(y_test, y_test_pred)
print(f"MSE: {mse:.8f}")
```

2. 並計算出測試階段的MSE

```python
plt.scatter(x_test, y_test, color='blue', label='Ground Truth')
plt.scatter(x_test, y_test_pred, color='red', label='Predict')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x)=1.6x+0.2,mse:0.41')
plt.show()  
```
